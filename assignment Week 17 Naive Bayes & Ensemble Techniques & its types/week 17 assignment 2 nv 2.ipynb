```python
# sol 1

# To find the probability that an employee is a smoker given that he/she uses the health insurance plan, we can use conditional probability. 
# P(A | B) = P(A ∩ B) / P(B).

    # The probability that an employee uses the insurance plan is 70% (0.70).
    # The probability that an employee who uses the insurance plan is a smoker is 40% (0.40).

# Now, we can calculate the conditional probability:

# P(Smoker | Uses Insurance Plan) = P(Smoker and Uses Insurance Plan) / P(Uses Insurance Plan) = 0.40 / 0.70 = 0.5714

# So, the probability that an employee is a smoker given that he/she uses the health insurance plan is approximately 0.5714 or 57.14%.
```

```python
# sol 2

# Bernoulli Naive Bayes and Multinomial Naive Bayes are two different variants of the Naive Bayes classifier, which is a probabilistic algorithm commonly used for text classification and other tasks. The main difference between them lies in the type of data they are designed to handle and how they model the data.

# 1. Bernoulli Naive Bayes:
    # Bernoulli Naive Bayes is typically used for binary classification problems, where the features are binary-valued (e.g., presence or absence of a feature).
    # It assumes that each feature is binary (0 or 1) and models the likelihood of each feature independently.
    # It's often used in text classification tasks where the presence or absence of words (features) in a document is used for classification.
    # Bernoulli Naive Bayes is not suitable for cases where we need to consider the frequency or count of features; it only considers whether a feature is present or not.

# 2. Multinomial Naive Bayes:
   # Multinomial Naive Bayes is designed for classification tasks with discrete data, typically text data where the features represent counts or frequencies.
   # It's commonly used in text classification problems, where the features are typically word counts or term frequencies (TF-IDF values).
   # It models the likelihood of each feature based on its frequency in the data, assuming that the features follow a multinomial distribution.
   # Multinomial Naive Bayes is well-suited for tasks where we want to consider the frequency or count of features, making it more flexible for text classification tasks.

# In summary, the choice between Bernoulli Naive Bayes and Multinomial Naive Bayes depends on the nature of the data we are working with. If we have binary features (presence/absence), Bernoulli Naive Bayes may be more appropriate. If we are dealing with text data and need to consider feature frequencies, Multinomial Naive Bayes is a better choice.
```

```python
corr_df=df.corr()
corr_df
# lets reduce some feturs
def reduce_correlation(dataset: pd.DataFrame, threshold: float):
    final = []
    corr = dataset.corr()
    for i in range(len(corr.columns)):
        for j in range(i):
            corr_no = corr.iloc[i, j]
            if corr_no >= threshold:
                final.append(corr.columns[i])
    return set(final)


reduce_correlation(df,0.83)
```

```python
df.drop(columns=['0.25', '0.31'],inplace=True)
```

```python
#split
y=df["1"]
x=df.drop(columns="1")

from sklearn.model_selection import train_test_split
x_train, X_test, y_train, Y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
## Implementation:
from sklearn.naive_bayes import BernoulliNB,MultinomialNB,GaussianNB
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
from sklearn.model_selection import cross_val_score

# for bernoulli

bernoulli_score=cross_val_score(BernoulliNB(),x_train,y_train,cv=10)
print("Bernoulli mean score = ",bernoulli_score.mean())

# for Multinomial

Multinomial_score=cross_val_score(MultinomialNB(),x_train,y_train,cv=10)
print("Multinomial mean score = ",Multinomial_score.mean())

# for Gaussian

Gaussian_score=cross_val_score(GaussianNB(),x_train,y_train,cv=10)
print("Gaussian mean score = ",Gaussian_score.mean())
```

```python
models={"bernoulli":BernoulliNB(),
        "Multinomial":MultinomialNB(),
        "Gaussian":GaussianNB()}

score_testing={accuracy_score.__name__:accuracy_score,
               precision_score.__name__:precision_score,
               recall_score.__name__:recall_score,
               f1_score.__name__:f1_score,
               }

def train_and_check(models_dict:dict,score_testing_dict:dict,x_train,y_train,X_test,Y_test):
    for model_name,model in models.items():
        print(f"\n\nThis is for {model_name}--\n")
        modelnaming=model.fit(x_train,y_train)
        for score_name,score in score_testing_dict.items():
            
            y_pred=modelnaming.predict(X_test)
            print(f"{score_name}={score(Y_test,y_pred)}")

train_and_check(models,score_testing,x_train,y_train,X_test,Y_test)
```
