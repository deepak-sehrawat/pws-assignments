```python
# sol 3

# DML (Data Manipulation Language) operations, such as INSERT, UPDATE, and DELETE, are typically performed on a database using SQL queries. In Python, you can interact with a database and execute these DML operations using database libraries and modules. 

# ***IMPORTANT this code is only working on vs code not in notbook ***

import mysql.connector
mydb = mysql.connector.connect(
  host="localhost",
  user="abc",
  password="password"
)

mycursor = mydb.cursor()
mycursor.execute("CREATE DATABASE if not exists my_db")
mycursor.execute("CREATE TABLE if not exists my_db.table1 (roll_no INT, first_name VARCHAR(12) , last_name VARCHAR(15), phone_no INT )")

# 1. INSERT:
        # The INSERT statement is used to add new data into a table. It allows you to specify the columns and the corresponding values to be inserted. 


mycursor.execute("INSERT INTO my_db.table1 values (19 , 'rohit' , 'kumar' ,74653874 )")
mycursor.execute("INSERT INTO my_db.table1 values (24 , 'deepak' , 'sehrawat' ,74687489 )")
mycursor.execute("INSERT INTO my_db.table1 values (67 , 'chander' , 'sehrawat' ,57866737 )")

# UPDATE: 
      # The UPDATE statement is used to modify existing data in a table. It allows you to specify which records to update and the new values for the specified columns. 

mycursor.execute("UPDATE my_db.table1 SET first_name = 'rohit' WHERE first_name='deepak'")

# DELETE:
        # The DELETE statement is used to remove specific records from a table. It allows you to specify the condition that identifies the records to be deleted.

mycursor.execute("DELETE FROM my_db.table1 WHERE first_name='chander'")


mydb.commit()
mydb.close()


```

```python
# sol 5
# A Primary Key :
            # is a column or a set of columns in a database table that uniquely identifies each record within the table. Its primary purpose is to ensure data integrity and enable efficient data retrieval.

# characteristics of a Primary Key:

# 1 Uniqueness: Each Primary Key value must be unique
# 2 Non-nullability: A Primary Key column cannot contain NULL values
# 3 Indexing: Primary Keys are often indexed, which allows for faster data retrieval and improves query performance.

# A Foreign Key 
            # establishes a relationship between two tables in a database. It is a column or a set of columns in one table that refers to the Primary Key of another table.

# characteristics of a Foreign key
# 1. Referential integrity: Foreign Keys enforce referential integrity by ensuring that the values in the referencing table match valid values in the referenced table's Primary Key.
# 2. Relationship establishment: Foreign Keys create relationships between tables, defining how data in one table relates to data in another.
#3. Cascading actions: Foreign Keys can define actions to be performed when a referenced record is updated or deleted, such as cascading updates or deletes.



```

```python
# sol 7
# clauses are  executed in the following order in sql:

# 1.FROM clause:
        # The FROM clause specifies the tables or views from which the data will be retrieved. 

# 2. WHERE clause:
        # The WHERE clause is used to filter the rows based on a specified condition. It is executed after the FROM clause and before the SELECT clause (if present). 

# 3. GROUP BY clause:
        # The GROUP BY clause is used to group the result set by one or more columns. It is executed after the WHERE clause and before the SELECT clause (if present). 

# 4. HAVING clause:
        # The HAVING clause is used to filter the groups created by the GROUP BY clause. It is similar to the WHERE clause, but it operates on the grouped data rather than individual rows.

# 5. SELECT clause:
        # The SELECT clause is used to specify the columns or expressions that will be included in the result set. 

# 6. ORDER BY clause:
        # The ORDER BY clause is used to sort the result set based on one or more columns. 

# 7. LIMIT 
        # The LIMIT (for databases like MySQL, PostgreSQL) or TOP (for databases like SQL Server) clause is used to restrict the number of rows returned by the query. 
```
