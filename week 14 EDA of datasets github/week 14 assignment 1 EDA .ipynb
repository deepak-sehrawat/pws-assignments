```python
# sol 2
df=pd.read_csv("winequality-red.csv")
df.isnull().sum()
## there is no null value in this dataset of redwine quality



# Handling missing data is a crucial step in the feature engineering process for any dataset, including our dataset the wine quality dataset. (but i don't get and null value in the dataset)
# Different imputation techniques have their advantages and disadvantages, and the choice of which one to use depends on the nature of the data and the goals of the analysis


# 1. Removing Rows with Missing Data (Listwise Deletion):

    # Advantages:
        # Simple and easy to implement.
        # Removes the problem of missing data entirely.

    # Disadvantages:
        # Can lead to a significant loss of data if there are many missing values.
        # May introduce bias if the missing data is not missing completely at random (MCAR).

# 2. Mean/Median Imputation:

    # Advantages:
        # Simple and does not change the data distribution significantly.
        # Useful for numerical features with missing values.

    # Disadvantages:
        # Ignores the relationships between features.
        # Can introduce bias, especially if data is not missing at random.
        # May not be suitable for categorical features.

# 3. Mode Imputation (for Categorical Data):

    # Advantages:
        # Appropriate for categorical variables.
        # Preserves the original distribution of the data for that feature.

    # Disadvantages:
        # May not be ideal if the mode is not representative of the true underlying distribution.
        # Like mean/median imputation, it does not consider relationships between variables.

# 4. Regression Imputation:

    # Advantages:
        # Considers relationships between variables by using regression models to predict missing values.
        # Preserves data distribution and relationships.

    # Disadvantages:
        # Requires more complex modeling and computation.
        # Sensitive to the quality of the regression model.
        # May not perform well with highly skewed or non-linear data.
```

```python
# sol 4 part 3 
from scipy.stats import shapiro

# Perform Shapiro-Wilk test for normality
for column in numeric_features:
    stat, p = shapiro(df[column])
    print(f"Shapiro-Wilk Test for {column}: p-value = {p}")
    if p < 0.05:
        print(f"{column} is not normally distributed.\n")
    else:
        print(f"{column} is normally distributed.\n")

# For features that exhibit non-normality, consider applying appropriate transformations to make the distribution more normal. Here are some common transformations we can use:

    # Logarithmic Transformation: Apply the natural logarithm (or log base 10) to the data if it is positively skewed.

    # Square Root Transformation: Use the square root transformation to reduce right-skewness.

    # Box-Cox Transformation: Apply the Box-Cox transformation to handle various types of skewness. This transformation automatically selects the best transformation parameter () to make the data as close to normal as possible.
    
    # Exponential Transformation: For negatively skewed data, consider taking the exponential of the data.
```

```python
# sol 6

df=pd.read_csv("winequality-red.csv")
from sklearn.preprocessing import StandardScaler

# Separate the target variable (quality) from the features
X = df.drop("quality", axis=1)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


from sklearn.decomposition import PCA

pca = PCA(n_components=None)
pca.fit(X_scaled)


explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance_ratio = explained_variance_ratio.cumsum()

# Find the number of components needed for 90% explained variance
n_components_90_percent = (cumulative_variance_ratio >= 0.90).sum() + 1
print(f"Number of components to explain 90% of variance: {n_components_90_percent}")

```
