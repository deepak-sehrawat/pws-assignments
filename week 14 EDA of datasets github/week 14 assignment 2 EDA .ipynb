```python
# sol 6

# To identify the peak travel season from the Flight Price dataset, we would need to analyze several key features. Here are the features to consider 

# Features to Analyze:

    # 1. Date/Time Information: The most crucial feature is the date of the flights. we should extract information such as the month, day of the week, and possibly the year.

# lets take our dataset to do the same 

df=pd.read_excel("flight_price.xlsx")
#fix the date formating 
# take Date_of_Journey column 
def fix_date(x):
    x=x.split("/")
    day,month,year=x[0],x[1],x[2]
    if len(day)==1:
        day="0"+day
    result=f"{year}-{month}-{day}"
    return result
df["date_of_journey_formated"]=df["Date_of_Journey"].apply(fix_date)

plt.figure(figsize=[15,10])
sns.countplot(hue=df["date_of_journey_formated"].str.split("-").str[1],x=df["date_of_journey_formated"].sort_values())
sns.histplot(x=df["date_of_journey_formated"].sort_values(),alpha=0.1,kde=True)
plt.xticks(rotation=90)
plt.show()

# from the given graph we can clearly se that in the dates 2019-06-06 to 2019-06-12 is the peak for the passanger to do the travaling 
#we show and give this information to our boss with this visual graph
```

```python
#sol 14 
# fix the install column  for calculataion 
def fix_install(x):
    x=str(x)
    re=x.replace(",","").replace("$","").replace("+","")
    return re

df["Installs"]=df["Installs"].apply(fix_install)
df_installs=df[["App","Installs"]]
df_installs["Installs"]=df_installs["Installs"].astype(int)

df_installs=df_installs.groupby("App")["Installs"].sum().reset_index()

df_installs.sort_values(by="Installs",ascending=False,ignore_index=True,inplace=True)
df_installs.head(10) ## that give the top 10 most popular apps by installs 
```

```python
# sol 15
## for this task we simply need two featurs app categories and no of installs 
df=pd.read_csv("googleplaystore.csv")

def fix_install(x):
    x=str(x)
    re=x.replace(",","").replace("$","").replace("+","")
    return re

df["Installs"]=df["Installs"].apply(fix_install)
df.drop(index=10472,axis=0,inplace=True)
df["Installs"]=df["Installs"].astype(int)

df_p_cat=df[["Category","Installs"]]
df_p_cat=df_p_cat.groupby("Category")["Installs"].sum().reset_index()

df_p_cat.sort_values(by="Installs",inplace=True,ascending=False,ignore_index=True)
print(df_p_cat) ## this gives the table of Category and there total no. of installs in sorted form 


## for our company we can make a bargarph to give them the more good Visualizations 

plt.figure(figsize=[15,7])
# sns.barplot(x=df_p_cat["Category"],y=df_p_cat["Installs"],palette="Set2")
sns.barplot(data=df_p_cat.head(10),x="Category",y="Installs",palette="Set2") ## this show the only bar graph of the top 10 most populer app categories 
plt.xticks(rotation=70)
plt.show()

# pie chart is also good  for this data Visualizations
plt.figure(figsize=[15,7])
df_p_cat_copy=df_p_cat[["Category","Installs"]][:10]
new_row=pd.DataFrame({"Category":["others"],"Installs":[df_p_cat["Installs"][10:].sum()]})
df_p_cat_copy2=pd.concat([df_p_cat_copy,new_row],axis=0)
plt.pie(data=df_p_cat_copy2,x="Installs",labels="Category",autopct='%1.1f%%',shadow=True)
plt.show()
```
