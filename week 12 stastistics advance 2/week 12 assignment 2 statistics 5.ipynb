```python
# importing of the important libraies 
import scipy.stats as stats
import math
import numpy as np 
```

```python
# sol 1
ci=95
x_mean=50
sd=5
si=1-ci/100
alfa=1-si/2
# let the n=1000 (sample size)
# we can use a large number like 10000 for a reasonable approximation.
sample_size = 1000
z_value=stats.norm.ppf(alfa).round(3)
range_value=(abs(z_value)*sd/sample_size**0.5)
range_value=round(range_value,3)
uper_range=x_mean+range_value
lower_range=x_mean-range_value
print("the range of mean for the 95% confidence interval is = ",lower_range,"-",uper_range)

# NOTE
# the accuracy of the confidence interval estimate depends on the sample size if the sample size is more the range get more narrow (it decrease the range of mean (it become more accurate) and if the sample size is get decrease the range of mean become more spread)
```

```python
# sol 2
exp_dis=[20,20,20,10,10,20]
# According to data from the Mars company, the color distribution of Milk Chocolate M&M's is approximately 24% blue, 20% orange, 16% green, 14% yellow, 13% red, and 13% brown.
obs_dis=[24,20,16,14,13,13]
# obs_dis=[24,20,16,10,10,20]
h0=" The obs distribution does not match the expected distribution"
h1=" The obs distribution  match the expected distribution"
sl=0.05
dof=len(exp_dis)-1
critical_value=stats.chi2.ppf(1-sl,dof)
chi_square,p_value=stats.chisquare(f_exp=exp_dis,f_obs=obs_dis)
if chi_square>critical_value:
    print ("we reject the null hypothesis -- the outcome is - ",h0)
else:
    print("we fail to reject the null hypothesis -- the outcome is  - ",h1 )
```

```python
# sol 3
import numpy as np
from scipy.stats import chi2_contingency

# Create a 2x2 contingency table
observed = np.array([[20, 15],
                     [10, 25],
                     [15, 20]])

# Perform the chi-square test
chi2, p,_,_ = chi2_contingency(observed)

print("Chi-square statistic:", chi2)
print("P-value:", p)

```

```python
# sol 4
n=500
smoker_propotion=65/500
cl=95
sl=1-cl/100
alfa=1-sl/2
z=stats.norm.ppf(alfa)
standard_error = math.sqrt((smoker_propotion * (1 - smoker_propotion))/n)
range_value=z*standard_error
uper_range=smoker_propotion+range_value
lower_range=smoker_propotion-range_value
# in percent 
uper_range=uper_range*100
lower_range=lower_range*100

print(f"the percentage range of smoker in the population is -- {lower_range:.3f}% to {uper_range:.3f}%")
```
