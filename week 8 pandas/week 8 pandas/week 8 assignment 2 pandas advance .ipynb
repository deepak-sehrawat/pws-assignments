```python
# sol 10 
import time
from datetime import timedelta,datetime
def update7(in_date):
    date=datetime.strptime(in_date,"%Y-%m-%d")
    new_date=date+timedelta(days=6)
    # datef=time.strftime("%Y-%m-%d",new_date)
    return new_date
    



df=pd.DataFrame({'Date':[
    "2023-02-14",
    "2023-04-05",
    "2023-06-22",
    "2023-07-11",
    "2023-08-27",
    "2023-09-18",
    "2023-10-29",
    "2023-11-16",
    "2023-12-03",
    "2023-12-28",
]})
ind=range(1,len(df))
df.reindex(ind)
df['MovingAverage']=df['Date'].apply(update7)
df
# print(time.strftime('%A',date))

```

```python
# sol 11
import time
def find_day(in_date):
    date=time.strptime(in_date,'%Y-%m-%d')
    day=time.strftime('%A',date)
    return day

df=pd.DataFrame({"Date":[
    "2023-02-14",
    "2023-04-05",
    "2023-06-22",
    "2023-07-11",
    "2023-08-27",
    "2023-09-18",
    "2023-10-29",
    "2023-11-16",
    "2023-12-03",
    "2023-12-28",
]})
df['Weekday']=df['Date'].apply(find_day)
print(df)
```

```python
# sol 12
df=pd.DataFrame({'Date':[
    "2023-02-14",
    "2023-01-05",
    "2023-06-22",
    "2023-07-11",
    "2023-08-27",
    "2023-09-18",
    "2023-01-29",
    "2023-11-16",
    "2023-12-03",
    "2023-12-28",
]})
def check_date(date):
    date=time.strptime(date,"%Y-%m-%d")
    sdate=time.strptime("2023-01-01","%Y-%m-%d")
    edate=time.strptime("2023-01-31","%Y-%m-%d")
    if date > sdate and date < edate:
        return True
    else:
        return False
df["condition"]=df["Date"].apply(check_date)
df[df["condition"]==True]

```
