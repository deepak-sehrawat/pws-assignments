```python
# sol 1

# A Support Vector Machine (SVM) is a supervised machine learning algorithm used for classification and regression tasks. In the case of a linear SVM for binary classification, the goal is to find a hyperplane that best separates two classes of data points. The mathematical formula for a linear SVM can be expressed as follows:

# Given a set of training data points (x_i, y_i), where x_i represents the feature vector of the ith data point, and y_i is the corresponding class label (+1 or -1 for binary classification), the objective of a linear SVM is to find the optimal weight vector (w) and bias term (b) that define the hyperplane. This hyperplane can be represented as:

# w * x + b = 0

    # Where:
    # - "w" is the weight vector (normal to the hyperplane).
    # - "x" is the feature vector of a data point.
    # - "b" is the bias term (an offset to move the hyperplane closer to one class or the other).

# The equation w * x + b = 0 defines the decision boundary. Data points on one side of this boundary will be classified as one class (e.g., +1), and those on the other side will be classified as the opposite class (e.g., -1).
```

```python
# sol 2

# The objective function of a linear Support Vector Machine (SVM) is to find a hyperplane that best separates two classes of data while maximizing the margin between them.

# This objective function is expressed as minimizing \({1}/{2}|w|^\), where (w) represents the weight vector of the hyperplane.
# Minimizing this term encourages finding a hyperplane with a larger margin, which is the distance between the hyperplane and the nearest data points from each class. 

# The constraints in the objective function, \(y_i(w \cdot x_i + b) \geq 1\), ensure that data points are correctly classified, with \(y_i\) as class labels and \(x_i\) as feature vectors. 

# The SVM seeks to strike a balance between maximizing the margin and minimizing classification errors. This optimization problem is typically solved using techniques like Sequential Minimal Optimization (SMO), resulting in the weight vector \(w\) and bias term \(b) that define the optimal hyperplane. This hyperplane is then used for classifying new data points based on the sign of (w * x + b).
```

```python
# sol 3

# The kernel trick is a crucial concept in Support Vector Machines (SVM) that enables these models to handle non-linearly separable data by implicitly mapping the input data into a higher-dimensional feature space. It allows SVM to find a hyperplane that can effectively separate classes that are not linearly separable in the original feature space.

# Here's how the kernel trick works:

    # 1. Mapping to a Higher-Dimensional Space:the kernel trick implicitly maps the data points to a higher-dimensional space where they might become linearly separable. This mapping is typically done through a mathematical function called a kernel. Common kernels include the polynomial kernel, radial basis function (RBF) kernel, and sigmoid kernel.

    # 2. Solving in the Higher-Dimensional Space: In this higher-dimensional space, SVM attempts to find a linear hyperplane that separates the classes. This is done using the same optimization techniques as a linear SVM, but now it operates in the transformed feature space.

    # 3. Prediction in the Original Space: After training, the SVM can classify new data points by mapping them into the same higher-dimensional space using the chosen kernel and then evaluating their position relative to the hyperplane in that space. The decision boundary in this higher-dimensional space corresponds to a non-linear decision boundary in the original feature space.

# The kernel trick allows SVMs to handle complex, non-linear relationships between features, making them powerful tools for a wide range of machine learning tasks, including image classification, text classification, and more. Common kernels and kernel parameters can be chosen based on the specific problem at hand.
```

```python
# sol 4

# Support vectors play a critical role in Support Vector Machines (SVM). They are the data points that are closest to the decision boundary (the hyperplane) and have a significant influence on the placement and orientation of that boundary. These support vectors help define the margin and contribute to the robustness of the SVM model.

# Here's an explanation with an example:

# Suppose we have a binary classification problem with two classes, "A" and "B," and we are using a linear SVM to separate them. The objective is to find the best hyperplane that maximizes the margin while correctly classifying the data points.

    # 1. Training Data: we have a dataset with several data points from both classes. For the sake of this example, let's consider three points from each class, but in practice, the dataset would be more extensive.

    #     Class A: (2, 2), (2, 3), (3, 3)
    #     Class B: (0, 0), (1, 0), (1, 1)

    # 2. Finding the Hyperplane: The SVM algorithm aims to find the hyperplane that separates the two classes while maximizing the margin. This hyperplane can be expressed as \(w*x + b = 0), where (w) is the weight vector and (b) is the bias term.

    # 3. Support Vectors: In this example, the support vectors are the data points closest to the decision boundary. In this case, they are:
    #     Support Vector 1: (2, 2) from Class A
    #     Support Vector 2: (1, 1) from Class B

    # 4. Margin: The margin of the SVM is defined by the distance between the decision boundary (hyperplane) and the support vectors. These support vectors are crucial because they determine the width of the margin. In the context of this example, the margin is determined by the distance between the hyperplane and the support vectors (2, 2) and (1, 1).


# support vectors are the data points that are closest to the decision boundary, and they play a crucial role in defining the margin and the orientation of the hyperplane. 
```

```python
# sol 5

# we can see all the lines or planes in the below image of graph which is created by matplotlib

# Hyperplane:
    # In SVM, the hyperplane is the decision boundary that separates the data into different classes. In a 2D feature space, it's a straight line.

    # Example: If we have two classes of data points, a hyperplane might be a line that separates these classes. For instance, in a binary classification problem with classes A and B, the hyperplane might be a line that best separates them.

# Margin:

    # The margin is the region between the support vectors and the hyperplane.
    # For a hard margin SVM, the margin is the space that should be as wide as possible while still separating the classes. It touches the support vectors.
    # For a soft margin SVM, the margin allows for some misclassification, and its width may vary depending on the data.

# Soft Margin vs. Hard Margin:

    # In a hard margin SVM, the margin is maximized while ensuring that no data point falls within the margin or on the wrong side of the hyperplane.
    # In a soft margin SVM, some data points are allowed to be inside the margin or on the wrong side of the hyperplane, and the margin may be narrower than in the hard margin case.
```

```python
import numpy as np

class LinearSVM:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, num_epochs=1000):
        self.learning_rate = learning_rate
        self.lambda_param = lambda_param
        self.num_epochs = num_epochs
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        num_samples, num_features = X.shape
        y_ = np.where(y <= 0, -1, 1)  # Convert class labels to -1 and 1

        # Initialize model parameters
        self.weights = np.zeros(num_features)
        self.bias = 0

        # Training the SVM using gradient descent
        for _ in range(self.num_epochs):
            for i in range(num_samples):
                if y_[i] * (np.dot(X[i], self.weights) - self.bias) >= 1:
                    self.weights -= self.learning_rate * (2 * self.lambda_param * self.weights)
                else:
                    self.weights -= self.learning_rate * (2 * self.lambda_param * self.weights - np.dot(X[i], y_[i]))
                    self.bias -= self.learning_rate * y_[i]

    def predict(self, X):
        linear_output = np.dot(X, self.weights) - self.bias
        return np.sign(linear_output)

if __name__ == '__main__':
    # Generate a synthetic dataset for testing
    np.random.seed(50)
    x = np.random.randn(200, 2)
    y = np.array([1] * 100 + [-1] * 100)
    x_train, X_test, y_train, Y_test = train_test_split(x, y, test_size=0.33, random_state=42)

    svm = LinearSVM(learning_rate=0.01, lambda_param=0.01, num_epochs=1000)
    # Make predictions 
    svm.fit(x_train, y_train)
    y_pred=svm.predict(X_test)
    from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
    print(f"{'x'*15} THIS IS THE RESULT OF MY SVM (WITHOUT SKLEARN) {'x'*15}\n")
    print("accuracy_score=",accuracy_score(Y_test,y_pred))
    print("classification_report=\n",classification_report(Y_test,y_pred))
    print("confusion_matrix=\n",confusion_matrix(Y_test,y_pred))


    from sklearn import svm
    vacter_machine=svm.SVC()
    vacter_machine.fit(x_train,y_train)
        
    # Make predictions using sklearn
    vacter_machine.fit(x_train, y_train)
    y_pred=vacter_machine.predict(X_test)
    
    from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
    print(f"\n\n{'x'*15} THIS IS THE RESULT OF SKLEARN SVM {'x'*15}\n")
    print("accuracy_score=",accuracy_score(Y_test,y_pred))
    print("classification_report=\n",classification_report(Y_test,y_pred))
    print("confusion_matrix=\n",confusion_matrix(Y_test,y_pred))


```
