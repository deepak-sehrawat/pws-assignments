```python
# sol 3

# app routing refers to the process of defining URL routes for different parts of a web application. 
#  It allows you to map specific URLs to corresponding functions in your Flask application.
# This mapping is achieved using decorators such as @app.route('/route_name').


# Usage

# URL Handling: App routes help in defining how URLs should be handled within the application. By mapping URLs to specific functions, we can determine the behavior of the application when a particular URL is accessed.

# Request Handling: App routes define the functions that handle incoming requests for specific URLs. When a user visits a particular URL, the associated function is executed, allowing you to process the request and generate a response.

# Code Organization: By using app routes, you can organize your codebase by separating different functionalities or pages into individual route handlers. 
```

```python
# sol 4
from flask import Flask
 
app=Flask(__name__)

@app.route("/welcome")
def welcome():
    return "<h3>Welcome to ABC Corporation</h3>"

@app.route("/")
def home():
    return "<div>Company name: ABC Corporation</div> <div>Location: India</div> <div>Contact Detail: 999-999-9999</div>"

app.run(host='0.0.0.0')
```
