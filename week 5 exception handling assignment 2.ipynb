```python
# sol 4 
# The LookupError class in Python is a base class for errors that occur when a lookup or indexing operation fails. It serves as the parent class for specific lookup-related errors. 

# two commonly encountered errors derived from the LookupError class: KeyError and IndexError.

# IndexError: This error occurs when trying to access a sequence like list,set,etc using an invalid index (out of range sequence ). 
#  example
try:
    my_list = [1, 2, 3]
    value = my_list[5]
    lg.info(f"{value}")
except LookupError as e:
    lg.error(f"Error 1 -- {e}")

# KeyError: This error is raised when trying to access a dictionary or mapping using a key that doesn't exist in the collection.
# example
try:
    my_dict = {'a': 1, 'b': 2, 'c': 3}
    value = my_dict["d"]
    lg.info(f"{value}")
except LookupError as e:
    lg.error(f"Error 2 -- key is missing {e}")

```

```python
# sol 5
# In Python, ImportError is an exception that is raised when an import statement fails to locate or load a module. It indicates a problem with importing and using external modules or packages. 

# ImportError: This error occurs when the import statement fails for any reason, such as the module not being found, errors in the module's code, or issues with dependencies

try :
    import try_to_import
except ImportError as e:
    lg.error(f"handling the emport error -- {e}")

# ModuleNotFoundError : This error is a subclass of ImportError . It was introduced to provide a more precise error message when a module is not found during the import process. 

try:
    from my_package import extra_module
except ModuleNotFoundError as e:
    lg.error(f"handling the module not found error -- {e}")
    


```
