```python
# sol 1
# Here are five types of plots that you can create using Seaborn along with their usage

# Scatter Plot:

    # Use: Scatter plots display the relationship between two continuous variables. Each point on the plot represents an observation in your dataset.

# Line Plot:

    # Use: Line plots (also known as line charts) are used to visualize the trend of a continuous variable over a range of values, often with respect to time or any other ordered variable.

# Bar Plot:

    # Use: Bar plots display categorical data using rectangular bars. Each bar's height or length represents the value of a category.

# Histogram:

    # Use: Histograms display the distribution of a single continuous variable by dividing it into bins and showing the frequency or count of data points within each bin.

# Heatmap:

    # Use: Heatmaps display a matrix of data as a grid of colored cells, where the colors represent the values of the matrix.


```
