```python
# sol 1
# Matplotlib is a popular Python library used for creating a wide variety of static, interactive, and animated visualizations in Python. 
# It provides a flexible and comprehensive framework for generating plots, charts, and graphs for data visualization and analysis.

# five plots that can be plotted using the Pyplot module of Matplotlib.

# 1 Line Plot: Used to visualize the relationship between two variables using lines connecting data points.

# 2 Bar Plot: Displays data with rectangular bars, where the height of each bar corresponds to the value of a variable.

# 3 Scatter Plot: Displays individual data points as dots on a two-dimensional plane, helpful for showing the correlation between two variables.

# 4 Histogram: Represents the distribution of data by grouping it into bins and displaying the frequency of occurrences in each bin.

# 5 Pie Chart: Displays data as slices of a circular pie, where each slice represents a proportion of the whole dataset.
```

```python
# sol 2
# A scatter plot is a type of data visualization that displays individual data points as dots on a two-dimensional plane.
# Each dot represents a single data instance characterized by two numeric variables. Scatter plots are used to visualize the relationship and correlation between these two variables. 
# The position of each dot on the plot is determined by its values on the X and Y axes, allowing for easy observation of patterns, trends, clusters, and potential outliers within the data. 
# Scatter plots are particularly useful for identifying associations and dependencies between variables and for understanding the dispersion or concentration of data points.

import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("this is our x-axis")
plt.ylabel("this is our y-axis")
plt.title("this is our graph")
```

```python

# sol 3
# The subplot() function is used to create a grid of subplots within a single figure in data visualization libraries like Matplotlib. 
# This function is particularly useful when you want to display multiple plots together in a single figure, allowing for easy comparison and visualization of various data or analyses side by side.
# the subplot() function enhances the clarity and effectiveness of data presentation by enabling multiple visualizations to coexist within a single figure.

import numpy as np 
import matplotlib.pyplot as plt

#for line 1 
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("line 1")

# for line 2
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("line 2")

# for line 3 
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title("line 3")

# for line 4
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title("line 4")

plt.tight_layout()# this is for better spacing 

plt.show()

```

```python
# sol 4 
# A bar plot is a type of data visualization that presents categorical data with rectangular bars, where the length of each bar corresponds to the value of a category. It is used to showcase comparisons, distributions, and relationships among discrete data sets. 

# usage-->
# Bar plots are particularly effective for displaying data with distinct categories or groups, allowing easy identification of trends, disparities, and rankings. 
# They simplify complex information and aid in making data-driven decisions by visually representing the magnitude of data points. 
# Bar plots are valuable tools in conveying insights to both technical and non-technical audiences, making data analysis and communication more accessible and impactful.




company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.xlabel("Company Name")
plt.ylabel("Profit")
plt.title(" Bar Graph of Profit ")


```

```python
# sol 5 
# A box plot, also known as a box-and-whisker plot, is a statistical data visualization that displays the distribution of a dataset along with its key summary statistics. 
# It consists of a rectangular "box" that represents the interquartile range (IQR) between the first quartile (Q1) and the third quartile (Q3), a horizontal line within the box representing the median, and "whiskers" extending to potential outliers.

# usage--->
# Box plots are used to depict the spread, central tendency, and presence of outliers in a dataset. They provide insights into the data's symmetry, skewness, and dispersion. By highlighting the median and quartiles, box plots offer a concise overview of the dataset's distribution, making it easier to identify variations and patterns within different categories or groups.
# Box plots are particularly useful in comparative analysis, showcasing differences between multiple datasets or conditions.
#Their simplicity and ability to represent statistical information make them valuable tools in exploratory data analysis, enabling efficient interpretation of complex data characteristics.

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1,box2])
```
